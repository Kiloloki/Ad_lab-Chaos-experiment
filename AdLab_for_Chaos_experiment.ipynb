```python
## 0.2304 g/s
import numpy as np
from scipy.stats import lognorm
import matplotlib.pyplot as plt
import pandas as pd
df = pd.read_csv("/content/Ad_lab_26/Data_5_Flowrate_0.2304.txt", sep=",", header=None)
T = df.iloc[:, 1].values

Tn = T[:-1]
Tn1 = T[1:]

plt.plot(Tn, Tn1, 'o')   # scatter-style
plt.xlabel("Tn")
plt.ylabel("Tn+1")
plt.show()

Tn_pos = Tn[Tn > 0]

# ---- Histogram as frequency (NOT density) ----
counts, bins, _ = plt.hist(Tn_pos, bins=100, density=False, alpha=0.6)

plt.xlabel("Tn (original)")
plt.ylabel("Frequency")

# ---- Fit lognormal ----
shape, loc, scale = lognorm.fit(Tn_pos, floc=0)

# ---- Convert PDF to frequency scale ----
x = np.linspace(min(Tn_pos), max(Tn_pos), 1000)
pdf = lognorm.pdf(x, shape, loc=loc, scale=scale)

N = len(Tn_pos)
bin_width = bins[1] - bins[0]

pdf_scaled = pdf * N * bin_width

plt.plot(x, pdf_scaled)

plt.show()
```

```python
import glob
import re

folder_path = "/content/Ad_lab_26/"
file_pattern = folder_path + "*Flowrate_*.txt"
file_paths = glob.glob(file_pattern)

all_flow_rates = []
all_Tn = []

flow_rate_for_std = []
std_values = []

for file in file_paths:
    match = re.search(r'Flowrate_(\d+\.\d+)', file)
    if not match:
        continue
    current_flow_rate = float(match.group(1))


    df = pd.read_csv(file, sep=",", header=None)
    T = df.iloc[:, 1].to_numpy()


    k_start = 5
    k_end = 5
    if len(T) <= k_start + k_end:
        continue
    T_trim = T[k_start : len(T)-k_end]


    all_flow_rates.extend([current_flow_rate] * len(T_trim))
    all_Tn.extend(T_trim)


    flow_rate_for_std.append(current_flow_rate)
    std_values.append(np.std(T_trim))


plt.figure(figsize=(12, 10))

plt.subplot(2, 1, 1)

plt.scatter(all_flow_rates, all_Tn, s=2, alpha=0.3, color='blue', marker='.')
plt.title("Bifurcation Diagram of Leaky Faucet")
plt.xlabel("Flow Rate (g/s)")
plt.ylabel("Time Interval $T_n$ (ms/s)")
plt.grid(True, linestyle='--', alpha=0.5)


plt.subplot(2, 1, 2)

sorted_indices = np.argsort(flow_rate_for_std)
sorted_flow_rates = np.array(flow_rate_for_std)[sorted_indices]
sorted_stds = np.array(std_values)[sorted_indices]

plt.plot(sorted_flow_rates, sorted_stds, 'r-o')
plt.title("Standard Deviation of $T_n$ vs Flow Rate")
plt.xlabel("Flow Rate (g/s)")
plt.ylabel("Standard Deviation of $T_n$")
plt.grid(True, linestyle='--', alpha=0.5)

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import glob
import re

folder_path = "/content/Ad_lab_26/"
file_pattern = folder_path + "*Flowrate_*.txt"
file_paths = glob.glob(file_pattern)

all_flow_rates = []
all_Tn = []
flow_rate_for_std = []
std_values = []

for file in file_paths:
    match = re.search(r'Flowrate_(\d+\.\d+)', file)
    if not match:
        continue
    current_flow_rate = float(match.group(1))

    df = pd.read_csv(file, sep=",", header=None)
    T = df.iloc[:, 1].to_numpy() / 1e6

    k_start = 5
    k_end = 5
    if len(T) <= k_start + k_end:
        continue
    T_trim = T[k_start : len(T)-k_end]

    all_flow_rates.extend([current_flow_rate] * len(T_trim))
    all_Tn.extend(T_trim)

    flow_rate_for_std.append(current_flow_rate)
    std_values.append(np.std(T_trim))


# plotting

plt.figure(figsize=(10, 8), dpi=150)

plt.subplot(2, 1, 1)
plt.scatter(all_flow_rates, all_Tn, s=1.5, alpha=0.4, color='#003366', marker='.')
plt.title("Bifurcation Diagram of a Leaky Faucet", fontsize=14)
plt.xlabel("Flow Rate (g/s)", fontsize=12)
plt.ylabel("Time Interval $T_n$ (s)", fontsize=12)
plt.grid(True, linestyle=':', alpha=0.7)


plt.subplot(2, 1, 2)
sorted_indices = np.argsort(flow_rate_for_std)
sorted_flow_rates = np.array(flow_rate_for_std)[sorted_indices]
sorted_stds = np.array(std_values)[sorted_indices]


plt.plot(sorted_flow_rates, sorted_stds, marker='o', linestyle='-', color='#CC0000', markersize=5, linewidth=1.5)
plt.title("Standard Deviation of $T_n$ vs Flow Rate", fontsize=14)
plt.xlabel("Flow Rate (g/s)", fontsize=12)
plt.ylabel("Standard Deviation of $T_n$ (s)", fontsize=12)
plt.grid(True, linestyle=':', alpha=0.7)

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.fft import fft, fftfreq
import glob
import re


def analyze_fft_for_flow_rate(folder_path, file_prefix, flow_rate_value, file_extension=".txt"):

    file_pattern = f"{folder_path}*Flowrate_*{flow_rate_value}*.txt"
    matched_files = glob.glob(file_pattern)

    if not matched_files:
        print(f"No {flow_rate_value:.4f}")
        return


    file_to_analyze = matched_files[0]
    print(f" {file_to_analyze} ...")


    df = pd.read_csv(file_to_analyze, sep=",", header=None)
    T = df.iloc[:, 1].to_numpy() / 1e6


    k_start = 5
    k_end = 5
    if len(T) <= k_start + k_end:
        print("No enough data")
        return
    T_trim = T[k_start : len(T)-k_end]


    T_detrended = T_trim - np.mean(T_trim)
    N = len(T_detrended)

    xf = fftfreq(N, 1)[:N//2]
    yf = fft(T_detrended)
    power_spectrum = 2.0/N * np.abs(yf[:N//2])

    plt.figure(figsize=(10, 6), dpi=120)


    plt.semilogy(xf, power_spectrum, color='#0066CC', linewidth=1.2)

    plt.title(f"FFT Power Spectrum (Flow Rate = {flow_rate_value:.4f} g/s)", fontsize=14)
    plt.xlabel("Frequency (drops$^{-1}$)", fontsize=12)
    plt.ylabel("Magnitude (Log Scale)", fontsize=12)
    plt.grid(True, which="both", linestyle=':', alpha=0.5)


    if flow_rate_value < 0.2:
        plt.axvline(x=0.5, color='r', linestyle='--', alpha=0.3, label="Period-2 Marker")
        plt.legend()

    plt.tight_layout()
    plt.show()


folder_path = "/content/Ad_lab_26/"
file_prefix = "Data_11_Flowrate_"


analyze_fft_for_flow_rate(folder_path, file_prefix, 0.1241)

print("\n")

analyze_fft_for_flow_rate(folder_path, file_prefix, 0.1433)
print("\n")
analyze_fft_for_flow_rate(folder_path, file_prefix, 0.2242)
print("\n")

analyze_fft_for_flow_rate(folder_path, file_prefix, 0.2795)

```

```python
def plot_3d_phase_space(folder_path, flow_rate_value):

    file_pattern = f"{folder_path}*Flowrate_*{flow_rate_value}*.txt"
    matched_files = glob.glob(file_pattern)

    if not matched_files:
        print(f"no {flow_rate_value} txt file。")
        return

    file_to_analyze = matched_files[0]
    print(f"Ploting{file_to_analyze} 3d graph...")

    df = pd.read_csv(file_to_analyze, sep=",", header=None)
    T = df.iloc[:, 1].to_numpy() / 1e6

    k_start = 5
    k_end = 5
    if len(T) <= k_start + k_end + 2:
        print("no enough data")
        return
    T_trim = T[k_start : len(T)-k_end]


    X = T_trim[:-2]      # T_n
    Y = T_trim[1:-1]     # T_{n+1}
    Z = T_trim[2:]       # T_{n+2}


    fig = plt.figure(figsize=(10, 8), dpi=120)
    ax = fig.add_subplot(111, projection='3d')


    time_color = np.arange(len(X))


    scatter = ax.scatter(X, Y, Z, c=time_color, cmap='viridis', s=8, alpha=0.8, edgecolors='none')


    ax.plot(X, Y, Z, color='gray', alpha=0.15, linewidth=0.5)

    ax.set_title(f"3D Phase Space Reconstruction (Flow Rate $\\approx$ {flow_rate_value} g/s)", fontsize=14)
    ax.set_xlabel("$T_n$ (s)", fontsize=12)
    ax.set_ylabel("$T_{n+1}$ (s)", fontsize=12)
    ax.set_zlabel("$T_{n+2}$ (s)", fontsize=12)


    cbar = plt.colorbar(scatter, ax=ax, pad=0.1, fraction=0.02)
    cbar.set_label('(Drop Index)', fontsize=10)


    ax.view_init(elev=20, azim=45)

    plt.tight_layout()
    plt.show()


folder_path = "/content/Ad_lab_26/"


plot_3d_phase_space(folder_path, 0.1433)

print("\n")
plot_3d_phase_space(folder_path, 0.1772)

print("\n")
plot_3d_phase_space(folder_path, 0.2795)
```
